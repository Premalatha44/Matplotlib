```python
# Stacked bar chart

plt.figure(figsize=(10,6))
plt.title("Stacked bar chart", fontsize=20, color="y")
labl=['a','b','c','d','e']
x=np.arange(len(labl))
bar1=[15,21,48,60,78]

bar2=[12,26,39,51,71]
bar3=[7,34,56,78,98]
plt.grid(axis='y', linestyle="-.", color='g', linewidth=1)
plt.bar(x,bar1)
plt.bar(x,bar2, bottom=bar1)
plt.bar(x,bar3, bottom=np.add(bar1,bar2))
plt.xticks(x, labl)
plt.show()
```
